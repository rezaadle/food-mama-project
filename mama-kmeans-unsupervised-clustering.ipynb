### Preparing a dataframe `cluster_data` for Unsupervised learning models

- Remove seasoning ingredients by their category_ids (sauces, épices, sel, sucres, beurre, jus, alcools, levure, eau).

```python
seasoning_categories = [18, 34, 37, 38, 39, 41, 42, 44, 45]
df = converted[converted['category_id'].apply(lambda x: x not in seasoning_categories)]
```

- Rearrange dataframe columns

```python
df = df.loc[:, ['title','recipe_id','origin_id','servings','name','food_id','category_id','availability','availability_cat']]
df.rename(columns={'title':'recipe_title', 'food_id':'ingredient_id', 'category_id':'ingredient_cat'}, inplace=True)
df.head(10)
```

 - Group dataframe by the number of ingredients used in each recipe and its availability

```python
#df.groupby(['recipe_title','origin_id','servings'])['ingredient_id','availability_cat'].count()
cluster_data = df.groupby(['recipe_id','origin_id','servings'])['ingredient_id','availability_cat'].count().reset_index()
cluster_data.loc[:,'availability_cat'] = [elm[1] for elm in df.groupby('recipe_id')['availability_cat'].value_counts().keys().tolist()]
cluster_data.rename(columns={'ingredient_id': 'ingredient_count'}, inplace=True)
```

### K-Means Clustering

```python
from sklearn.cluster import KMeans

# Drop "recipe_id" column as it brings no value to our model
X = cluster_data.drop(['recipe_id'], axis=1)
```

#### The features in the dataset contain different ranges of values. So, it's important to transform them to the same range.
#### 1) Scaling the Data

```python
from sklearn.preprocessing import MinMaxScaler
X_scaled = MinMaxScaler().fit_transform(X)
```

#### 2) Standardize the Data

```python
from sklearn.preprocessing import StandardScaler
X_stand = StandardScaler().fit_transform(X)
```

### Two Elbow methods, to find optimal number of clusters

```python
# Euclidean distance
from scipy.spatial.distance import cdist
clusters = range(1,15)
meanDistortions = []

for k in clusters:
    model_1 = KMeans(n_clusters=k)
    model_1.fit(X_scaled)
    prediction = model_1.predict(X_scaled)
    meanDistortions.append( sum(np.min(cdist(X_scaled, model_1.cluster_centers_, 'euclidean'), axis=1)) / X_scaled.shape[0] )
    
plt.plot(clusters, meanDistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
plt.show()
```

```python
# Sum of squared distances
sse = []
K = range(1,15)

for k in K:
    model_2 = KMeans(n_clusters=k)
    model_2 = model_2.fit(X_stand)
    sse.append(model_2.inertia_)
    
plt.plot(K, sse, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Optimal clusters is 6
final_model = KMeans(n_clusters=6, random_state=0)
final_model.fit(X_scaled)
prediction = final_model.predict(X_scaled)
 
# Join predicted clusters to final_data
final_table = cluster_data.copy()
final_table["Cluster"] = prediction
final_table.head()
```

```python
import seaborn as sb
sb.pairplot(final_table, hue='Cluster')
```

### Ploting with centroids

```python
# Set the final_data that will be used
final_data = X_stand
cluster = 4

final_km = KMeans(algorithm='auto', n_clusters=cluster, random_state=0)
final_km.fit(final_data)
labels = final_km.predict(final_data)
centroids = final_km.cluster_centers_
```
