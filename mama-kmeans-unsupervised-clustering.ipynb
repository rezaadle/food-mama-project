### K-Means Clustering

```python
from sklearn.cluster import KMeans
# Drop "recipe_id" column as it brings no value to our models
X = cluster_data.drop(['recipe_id'], axis=1)
```

#### The features in the dataset contain different ranges of values. So, it's important to transform them to the same range.
#### 1) Scaling the Data

```python
from sklearn.preprocessing import MinMaxScaler
X_scaled = MinMaxScaler().fit_transform(X)
```

#### 2) Standardize the Data

```python
from sklearn.preprocessing import StandardScaler
X_stand = StandardScaler().fit_transform(X)
```

### Two Elbow methods, to find optimal number of clusters

```python
# Euclidean distance
from scipy.spatial.distance import cdist
clusters = range(1,15)
meanDistortions = []

for k in clusters:
    model_1 = KMeans(n_clusters=k, random_state=0)
    model_1.fit(X_scaled)
    prediction = model_1.predict(X_scaled)
    meanDistortions.append( sum(np.min(cdist(X_scaled, model_1.cluster_centers_, 'euclidean'), axis=1)) / X_scaled.shape[0] )
    
plt.plot(clusters, meanDistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
plt.show()
```

```python
# Sum of squared distances
sse = []
K = range(1,15)

for k in K:
    model_2 = KMeans(n_clusters=k, random_state=0)
    model_2 = model_2.fit(X_stand)
    sse.append(model_2.inertia_)
    
plt.plot(K, sse, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method for Optimal k')
plt.show()
```

- Optimal number of clusters is 9

```python
final_model = KMeans(n_clusters=9, random_state=0)
final_model.fit(X_scaled)
prediction = final_model.predict(X_scaled)
 
# Join predicted clusters to final_data
final_table = cluster_data.copy()
final_table["Cluster"] = prediction
final_table.head()
```

```python
sb.pairplot(final_table.iloc[:, [0,1,2,3,4,-1]], hue='Cluster')
```

### Ploting with centroids

```python
# Set the final_data that will be used
final_data = X_scaled
cluster = 4

final_km = KMeans(algorithm='auto', n_clusters=cluster, random_state=0)
final_km.fit(final_data)
labels = final_km.predict(final_data)
centroids = final_km.cluster_centers_
```

```python
plt.scatter(final_data[:, 0], final_data[:, 3], c=labels, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 3], c='red', s=200, alpha=0.7)
plt.show()
```
